## Starter Code to Import Libraries and Load the Weather and Coordinates Data

```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
import os
import json
# Import API key
from api_keys import geoapify_key
```

```python
# Load the CSV file created in Part 1 into a Pandas DataFrame
city_data_df = pd.read_csv("output_data/cities.csv") 

# Display sample data
city_data_df.head()
```

---

### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
%%capture --no-display
# Configure the map plot
map_plot = city_data_df.hvplot.points(
    'Lng',
    'Lat',
    tiles = 'OSM',
    coastline = True,
    frame_width = 800,
    frame_height = 500,
    hover_cols = ['City', 'Country', 'Lat', 'Lng'],
    size = 'Humidity'
)

# Display the map
map_plot
```

## Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Narrow down cities that fit criteria and drop any results with null value 
cleaned_city_data_df = city_data_df.loc[(city_data_df["Max Temp"] > 20) & (city_data_df["Max Temp"] < 26) & (city_data_df["Cloudiness"] == 0)]
# Drop any rows with null values
cleaned_city_data_df = cleaned_city_data_df.dropna(how='any') 
# Display sample data
cleaned_city_data_df
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = cleaned_city_data_df[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df['Hotel Name'] = ''

# Display sample data
hotel_df
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
# Set parameters to search for a hotel
radius = 10000

params = {
    'apiKey': geoapify_key,
}

# Print a message to follow up the hotel search
print("Starting hotel search\n---------------------")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # get latitude, longitude from the DataFrame
    latitude = hotel_df.loc[index, 'Lat']
    longitude = hotel_df.loc[index, 'Lng']
    
    # set filters, bias, categories, and limit
    filters = f'circle:{longitude},{latitude},{radius}'
    bias = f'proximity:{longitude},{latitude}'
    categories = 'accommodation.hotel'
    limit = 1
    
    # Add filter and bias parameters with the current city's latitude and longitude to the params dictionary
    params['filter'] = filters
    params['bias'] = bias
    params['categories'] = categories
    params['limit'] = limit    
    
    # Set base URL
    base_url = 'https://api.geoapify.com/v2/places'

    # Make and API request using the params dictionaty
    name_address = requests.get(base_url, params = params)
    
    # Convert the API response to JSON format
    name_address = name_address.json()
    
    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"
        
    # Log the search results
    print(f"{hotel_df.loc[index, 'City']}, {hotel_df.loc[index, 'Country']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Print a message to confirm the hotel search end
print("---------------------\nHotel search completed\n")
    
# Display sample data
hotel_df
    
```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
%%capture --no-display

# Configure the map plot
map_plot_hotel = hotel_df.hvplot.points(
    'Lng',
    'Lat',
    geo = True,
    tiles = 'OSM',
    coastline = True,
    frame_width = 800,
    frame_height = 500,
    hover_cols = ['City', 'Country', 'Lat', 'Lng', 'Hotel Name'],
    size = 'Humidity'
)

# Display the map
map_plot_hotel
```
